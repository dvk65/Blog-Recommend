```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
blogs = pd.read_excel('Blogs_Combined.xlsx')
blogs.shape
```

```python
blogs.head()
```

```python
#tfidf converts text to numbers. number of features=number of distinct words in overview column. 
#values directly proportional to number of times a word is used and inversely proportional to the number of documents (movies).
#Words that occur multiple times but are common to many movies may not be useful (as, a, the, an, are, she, is, he)

tfidf = TfidfVectorizer(stop_words='english')
blogs['info']=blogs['info'].fillna('')

#"fit" computes the mean and std to be used for later scaling. (just a computation), nothing is given to you. 
#"transform" uses a previously computed mean and std to autoscale the data (subtract mean from all values and then divide it 
#by std). "fit_transform" does both at the same time.

info_matrix=tfidf.fit_transform(blogs['info'])
info_matrix.shape
```

```python
#kernel are used to separate non linear data. Unlike support vector machines which form clusters separated by a linear equation
similarity_matrix=linear_kernel(info_matrix,info_matrix)
similarity_matrix
```

```python
#cos angle between the vectors calculated. Smaller cos angle is more similar
similarity_matrix = cosine_similarity(info_matrix,info_matrix)
```

```python
#series convert a fixed size 1D array (ndarray) to list with index values
mapping = pd.Series(blogs.index,index = blogs['title'])
mapping
```

```python
def recommend_blogs(blog_title):
    blog_index = mapping[blog_title]
    #enumerate helps with managing content for iterating tools
    similarity_list=list(enumerate(similarity_matrix[blog_index]))
    similarity_list=sorted(similarity_list, key=lambda x:x[1], reverse=True)
    similarity_list=similarity_list[1:15]
    blog_index = [i[0] for i in similarity_list]
    return (blogs['title'].iloc[blog_index])
```

```python
#linear_kernel tfidf
recommend_blogs('Canada : Get To Know Your Dream Destination')
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('averaged_perceptron_tagger')
from nltk.corpus import wordnet
  
lemmatizer = WordNetLemmatizer()
  
# Define function to lemmatize each word with its POS tag
  
def pos_tagger(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:          
        return None

#sentence = 'the cat is sitting with the bats on the striped mat under many badly flying geese'
  
# tokenize the sentence and find the POS tag for each token
lemmatized_sentence_df = []
for bi in blogs['info']:
    pos_tagged = nltk.pos_tag(nltk.word_tokenize(bi))  

    print(pos_tagged)

    # As you may have noticed, the above pos tags are a little confusing.
    # we use our own pos_tagger function to make things simpler to understand.
    wordnet_tagged = list(map(lambda x: (x[0], pos_tagger(x[1])), pos_tagged))
    print(wordnet_tagged)

    lemmatized_sentence=[]
    for word, tag in wordnet_tagged:
        if tag is None:
            # if there is no available tag, append the token as is
            lemmatized_sentence.append(word)
        else:  
            # else use the tag to lemmatize the token
            lemmatized_sentence.append(lemmatizer.lemmatize(word, tag))
    lemmatized_sentence = " ".join(lemmatized_sentence)
    lemmatized_sentence_df.append(lemmatized_sentence)
blogs['info']=lemmatized_sentence    
```
